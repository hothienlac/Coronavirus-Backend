```python
import pandas as pd
```

```python
confirmed = pd.read_csv('confirmed.csv')
```

```python
x= confirmed
```

```python
confirmed = x
```

```python
confirmed = pd.concat([confirmed.iloc[:,1].rename('Country'), confirmed.iloc[:,-1].rename('cases')], axis=1)
confirmed
```

```python
confirmed = confirmed.groupby('Country').sum()
confirmed
```

```python
recovered = pd.read_csv('recovered.csv')
```

```python
recovered = pd.concat([recovered.iloc[:,1].rename('Country'), recovered.iloc[:,-1].rename('recovered')], axis=1)
recovered
```

```python
recovered = recovered.groupby('Country').sum()
recovered
```

```python
death = pd.read_csv('death.csv')
death = pd.concat([death.iloc[:,1].rename('Country'), death.iloc[:,-1].rename('death')], axis=1)
death = death.groupby('Country').sum()
death
```

```python
result = pd.concat([confirmed, recovered, death], axis=1, sort=False)
```

```python
result = result.reset_index()
```

```python
result['flag'] = 'assets/flags/' + result['Country'] + '.png'
result
```

```python
result['active'] = result['cases'] - result['death'] - result['recovered']
result
```

```python
result['death_rate'] = round(result['death'] / result['cases']*100, 2)
result
```

```python
result['recovered_rate'] = round(result['recovered'] / result['cases']*100, 2)
result
```

```python
result = result.sort_values('cases', axis=0, ascending=False)
```
