```python
import pandas as pd

from cachetools import cached, TTLCache

cache1 = TTLCache(maxsize=1, ttl=100)
cache2 = TTLCache(maxsize=1, ttl=100)
cache3 = TTLCache(maxsize=1, ttl=100)

@cached(cache1)
def getConfirmed():
    data = pd.read_csv('confirmed.csv')
    return data

@cached(cache2)
def getRecovered():
    data = pd.read_csv('recovered.csv')
    return data

@cached(cache3)
def getDeath():
    data = pd.read_csv('death.csv')
    return data
```

```python
confirmed = getConfirmed()
confirmed = confirmed.iloc[:,4:]
total_confirmed = confirmed.sum(axis=0)

recovered = getRecovered()
recovered = recovered.iloc[:,4:]
total_recovered = recovered.sum(axis=0)

death = getDeath()
death = death.iloc[:,4:]
total_death = death.sum(axis=0)

date = total_confirmed.index[1:].values.tolist()
```
